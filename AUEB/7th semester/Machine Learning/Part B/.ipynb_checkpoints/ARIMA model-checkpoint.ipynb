```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize']=20,10
from statsmodels.tsa.arima.model import ARIMA # arima import gia xrhsh tou paketou


dataset = pd.read_csv('BTC-USD.csv')
dataset.head()
```

```python
dataset["Date"]=pd.to_datetime(dataset.Date,format = "%d/%m/%Y")
dataset.index = dataset['Date']

plt.figure(figsize=(16,8))
plt.plot(dataset["Close"],label='Close Price history')   # arxiko diagramma twn dedomenwn
plt.legend(loc='upper left', fontsize=15)
plt.show()
```

```python
# xekiname me q = 1

train_data = dataset[:1461]['Close'].values # 80% training data
test_data = dataset[1461:]['Close'].values # 20% testing data


history = [x for x in train_data]
predictions = list()

"""
for i in range(len(test_data)):
    model = ARIMA(history, order=(1,1,2))
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test_data[i]
    history.append(obs)
    #print('predicted=%f, expected=%f' % (yhat, obs))

plt.plot(test_data , label = 'Actual Close values')
plt.plot(predictions , label = 'Predicted Close values')

"""


model = ARIMA(dataset['Close'], order=(0,0,1)) # mono MA(q) = 1 , ta alla 0.
model_log = model.fit()
print(model_log.summary())

# provlepsh

dataset["Predictions"] = model_log.predict(start = "13-2-2021" ,end = "14-12-2021" ,dynamic=False)
plt.plot(dataset["Close"], label = "Actual values")
plt.plot(dataset["Predictions"] , label = "Predicted values")
plt.legend(loc='upper left', fontsize=15)
plt.show()
```
