# Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('BTC-USD.csv')
data.head()

x=data['Date']
y=data['Close']

print(x.head())
print(y.head())


```

```python
fig1=plt.figure(figsize=(8,5))
ax1=fig1.add_axes([0,0,1,1])
ax1.set_xlabel('Date')
ax1.set_ylabel('Closing Price')
ax1.set_title('BTC closing price from ...to....')
ax1.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.xticks(rotation=45)
ax1.plot(x,y, color='orange')
```

```python
fig2,ax2=plt.subplots(1,1,figsize=(15,6))
plt.fill_between(x,y1=y,y2=-y, color='blue')
plt.ylim(-70000,70000)
ax2.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.xticks(rotation=45)
```

```python
data=pd.read_csv('BTC-USD.csv',parse_dates=["Date"])
#print(data.head())

#convert dates from string to datetime objects
data["Date"]=pd.to_datetime(data["Date"])

#keep only the date part of datetime object
data["Date"]=data["Date"].dt.date


```

```python
data["Year"]=[data_point.year for data_point in data["Date"]]
data["Year"].head()
data["Month"]=[data_point.month for data_point in data["Date"]]
data["Day"]=[data_point.day for data_point in data["Date"]]


```

```python
fig3=plt.figure(figsize=(14,10))
axes3=fig3.add_axes([0,0,1,1])

for i, y in enumerate(data["Year"].unique()):
    axes3.plot(data.loc[data["Year"]==y]["Close"].groupby(data["Month"]).mean(), label=y)
   

axes3.legend(prop={'size':15})
axes3.set_xlabel("Month")
axes3.set_ylabel("BTC Price")
axes3.set_title("Seasonal price fluctuations")


```
